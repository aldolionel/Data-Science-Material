1. Calculate the number of missing values in each column of the dummy DataFrame.
2. Calculate the range of the column.
3. Calculate random values with the size of num_nulls.
4. In your function definition, set the default value of scaling_factor to be 0.075
5. Next, scale your dummy values by scaling_factor and multiply them by col_range. The minimum col.min() has been already added for you.

```python
def fill_dummy_values(df, scaling_factor=0.075):
  df_dummy = df.copy(deep=True)
  for col_name in df_dummy:
    col = df_dummy[col_name]
    col_null = col.isnull()    
    # Calculate number of missing values in column 
    num_nulls = col_null.sum()
    # Calculate column range
    col_range = col.max() - col.min()
    # Scale the random values to scaling_factor times col_range
    dummy_values = (rand(num_nulls) - 2) * scaling_factor * col_range + col.min()
    col[col_null] = dummy_values
  return df_dummy
```

1. Fill dummy values in diabetes_dummy using fill_dummy_values().
2. Sum the nullity of 'BMI' and 'Skin_Fold'.
3. Create a scatter plot of 'Skin_Fold' vs. 'BMI'.

**HINT!**

* Set BMI as y-axis and Skin_Fold as x-axis.
* To sum the nullity of two columns, you can add both their missing values generated from the .isnull() method.
* The c argument controls the color of the points in the plot.

```python
from numpy.random import rand

# Fill dummy values in diabetes_dummy
diabetes_dummy = fill_dummy_values(diabetes)

# Sum the nullity of Skin_Fold and BMI
nullity = diabetes['Skin_Fold'].isnull()+diabetes['BMI'].isnull()

# Create a scatter plot of Skin Fold and BMI 
diabetes_dummy.plot(x='Skin_Fold', y='BMI', kind='scatter', alpha=0.5,
                    
                    # Set color to nullity of BMI and Skin_Fold
                    c=nullity, 
                    cmap='rainbow')

plt.show()
```
